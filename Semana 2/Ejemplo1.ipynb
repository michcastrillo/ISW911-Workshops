```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
with open('/content/sample_data/listado.txt', 'r', encoding='utf-8') as file:
    data = file.read()
data = data.replace(';', '')
data = data.replace('de', '')
wordcloud = WordCloud(width=900, height=400, background_color='white').generate(data)
word_counts = Counter(data.split())
top_words = word_counts.most_common(30)
plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Nube de Palabras')
# Grfico de barras
plt.subplot(122)
words, counts = zip(*top_words)
bars = plt.barh(words, counts)
plt.xlabel('Frecuencia')
plt.ylabel('Palabras')
plt.title('Palabras ms Frecuentes')
# Agregar etiquetas en las barras
for bar, count in zip(bars, counts):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, count, ha='left', va='center')
plt.tight_layout()
plt.show()
```
